```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import numpy as np

X = df[["sepal_length","sepal_width","petal_length","petal_width"]].values
y = df[["species"]].values
print(f"Son {X.shape[0]} muestras totales")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)
print(f"Son {X_train.shape[0]} muestras para entrenamiento y {X_test.shape[0]} datos para prueba")
```

```python
y_test
```

```python
algoritmo = KNeighborsClassifier(n_neighbors=5)
algoritmo.fit(X_train, y_train)

Y_pred = algoritmo.predict(X_test)
accuracy = algoritmo.score(X_train, y_train)
print(f"Exactitud Vecinos m치s Cercanos: {accuracy*100}%")
```

```python
# Graficando muestras de entrenamiento
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)

fig, ax = plt.subplots(figsize=(6.8,4.2))
# X_train[:,-2:]
# y_train
species = np.unique(y_train)
y_train[y_train == species[0]] = 0
y_train[y_train == species[1]] = 1
y_train[y_train == species[2]] = 2
y_train = y_train.flatten()

X_train = X_train[:,-2:]

colors = ["tab:blue", "tab:orange", "tab:green"]

for k in range(len(species)):
    i_species = y_train == k
    ax.scatter(X_train[i_species,0], X_train[i_species,1], c=colors[k], label=species[k])
    
ax.set_ylabel("petal_width")
ax.set_xlabel("petal_length")
ax.legend()
```

```python
# Graficando muestras de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)

fig, ax = plt.subplots(figsize=(6.8,4.2))

species = np.unique(y_train)
y_train[y_train == species[0]] = 0
y_train[y_train == species[1]] = 1
y_train[y_train == species[2]] = 2
y_train = y_train.flatten()

y_test[y_test == species[0]] = 0
y_test[y_test == species[1]] = 1
y_test[y_test == species[2]] = 2
y_test = y_test.flatten()

X_train = X_train[:,-2:]

X_test = X_test[:,-2:]

colors = ["tab:blue", "tab:orange", "tab:green"]

for k in range(len(species)):
    # Entrenamiento
    i_species = y_train == k
    ax.scatter(X_train[i_species,0], X_train[i_species,1], c=colors[k], label=species[k])
    # Pruebas
    i_species = y_test == k
    ax.scatter(X_test[i_species,0], X_test[i_species,1], c=colors[k], marker="d", edgecolors="k", zorder=5)
    
ax.set_ylabel("petal_width")
ax.set_xlabel("petal_length")
ax.legend()
```

```python
# Graficando muestras de entrenamiento, prueba y nuevas muestras
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=30)

fig, ax = plt.subplots(figsize=(6.8,4.2))

species = np.unique(y_train)
y_train[y_train == species[0]] = 0
y_train[y_train == species[1]] = 1
y_train[y_train == species[2]] = 2
y_train = y_train.flatten()

y_test[y_test == species[0]] = 0
y_test[y_test == species[1]] = 1
y_test[y_test == species[2]] = 2
y_test = y_test.flatten()

y_pred = iris["species"].values
y_pred[y_pred == species[0]] = 0
y_pred[y_pred == species[1]] = 1
y_pred[y_pred == species[2]] = 2
y_pred = y_pred.flatten()

X_train = X_train[:,-2:]

X_test = X_test[:,-2:]

X_pred = iris[["petal_length","petal_width"]].values

colors = ["tab:blue", "tab:orange", "tab:green"]

for k in range(len(species)):
    # Entrenamiento
    i_species = y_train == k
    ax.scatter(X_train[i_species,0], X_train[i_species,1], c=colors[k], label=species[k])
    # Pruebas
    i_species = y_test == k
    ax.scatter(X_test[i_species,0], X_test[i_species,1], c=colors[k], zorder=5)
    # Nuevas muestras
    i_species = y_pred == k
    ax.scatter(X_pred[i_species,0], X_pred[i_species,1], c=colors[k], marker="s", zorder=10, edgecolors="k")
    
ax.set_ylabel("petal_width")
ax.set_xlabel("petal_length")
ax.legend().set_zorder(50)
```
